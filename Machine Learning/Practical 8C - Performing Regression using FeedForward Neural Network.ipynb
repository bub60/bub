Practical 2: Performing Regression using FeedForward Neural Network

```python
import numpy as np
```

```python
np.random.seed(42)
```

```python
X = np.random.rand(1000,10)
```

```python
y = np.random.rand(1000)
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(12,activation='relu',input_dim=10))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='linear'))
```

```python
model.compile(loss='mse',optimizer='adam',metrics=['mae'])
```

```python
model.fit(xtrain,ytrain,epochs=400,batch_size=50)
```

```python
predictions = model.predict(xtest)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
r2_score(ytest,predictions)
```

```python
mean_absolute_error(ytest,predictions)
```
